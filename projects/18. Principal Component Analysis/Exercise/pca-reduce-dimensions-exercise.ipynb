```python
df = pd.read_csv("/kaggle/input/heart-csv/heart.csv")

df.head()
```

```python
df.shape
```

## DATA PROCESSING

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df['Sex'] = df.Sex.map({"M": 0, "F": 1})
df.Sex
```

```python
df.ChestPainType.value_counts()
```

```python
dummies = pd.get_dummies(df.ChestPainType).astype(int)
dummies.head()
```

```python
df = df.drop(columns = ['ChestPainType'])
df.head()
```

```python
df = pd.concat([df, dummies], axis = 1)
df.head()
```

```python
df.dtypes
```

```python
df.RestingECG.value_counts()

dum = pd.get_dummies(df.RestingECG).astype(int)
df = df.drop(columns = ['RestingECG'])
df = pd.concat([df, dum], axis = 1)

df.head()
```

```python
df.dtypes
```

```python
print(df.ExerciseAngina.value_counts())

df['ExerciseAngina'] = df['ExerciseAngina'].map({"N": 0, "Y": 1})
df.head()
```

```python
print(df.ST_Slope.value_counts())

dumm = pd.get_dummies(df.ST_Slope).astype(int)
print(dumm.head())

df = df.drop(columns = ['ST_Slope'])
df = pd.concat([df, dumm], axis = 1)

df.head()
```

```python
df.dtypes
```

```python
X = df.drop(columns = ['HeartDisease'])
y = df['HeartDisease']

X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter = 1000)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
X.shape
```

```python
from sklearn.decomposition import PCA

pca = PCA(0.95)
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
from sklearn.model_selection import train_test_split

X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter = 1000)
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 10)
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
from sklearn.model_selection import train_test_split

X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size = 0.3, random_state = 42)

from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter = 1000)
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
```
