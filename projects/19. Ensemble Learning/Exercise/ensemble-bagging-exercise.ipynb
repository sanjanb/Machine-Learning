## IMPORT LIBRARIES AND DATASET

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("/kaggle/input/enseble-heart-failure-dataset/heart.csv")

df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df['Sex'] = df.Sex.map({"M": 0, "F": 1})
df.Sex
```

```python
print(df['ChestPainType'].value_counts())

dummies = pd.get_dummies(df['ChestPainType']).astype(int)

df.drop(columns = ['ChestPainType'], inplace = True)
df = pd.concat([df, dummies], axis = 1)

df.head()
```

```python
df.dtypes
```

```python
print(df['RestingECG'].value_counts())

dum = pd.get_dummies(df['RestingECG']).astype(int)

df.drop(columns = ['RestingECG'], inplace = True)

df = pd.concat([df, dum], axis = 1)

df.head()
```

```python
df.dtypes
```

```python
df['ExerciseAngina'] = df.ExerciseAngina.map({'N': 0, 'Y': 1})
df['ExerciseAngina']
```

```python
print(df['ST_Slope'].value_counts())

dumm = pd.get_dummies(df['ST_Slope']).astype(int)
df.drop(columns = ['ST_Slope'], inplace = True)

df = pd.concat([df, dumm], axis = 1)

df.head()
```

```python
df.dtypes
```

```python
X = df.drop(columns = ['HeartDisease'])
y = df['HeartDisease']

X.shape, y.shape
```

```python
from sklearn.preprocessing import StandardScaler

s = StandardScaler()
X_scaled = s.fit_transform(X)

X_scaled
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC

scores = cross_val_score(SVC(), X_scaled, y, cv = 5)
np.average(scores)
```

```python
from sklearn.ensemble import BaggingClassifier

model = BaggingClassifier(
    estimator = SVC(),
    n_estimators = 100,
    max_samples = 0.8, 
    oob_score = True,
    random_state = 42
)

model.fit(X_train, y_train)
model.oob_score_
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

score = cross_val_score(RandomForestClassifier(), X_scaled, y, cv = 5)
np.average(score)
```
