```python
import pandas as pd
import numpy as np
import math
from sklearn import linear_model
```

```python
df = pd.read_csv("/kaggle/input/homeprices-csv/homeprices.csv")
```

```python
df.head()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

## SAVE THE MODEL USING PICKLE

```python
import pickle
```

```python
# save the python object

with open('model_pickle', 'wb') as f:
    pickle.dump(reg, f)
```

```python
with open('model_pickle', 'rb') as f:
    model = pickle.load(f)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict(np.array([5000]).reshape(1, -1))
```

## SAVE THE MODEL USING SKLEARN'S JOBLIB
